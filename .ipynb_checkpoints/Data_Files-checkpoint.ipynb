```python
import tensorflow.compat.v1 as tf
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
(x_train,y_train),(x_test,y_test)=tf.keras.datasets.boston_housing.load_data(path='boston_housing.npz',test_split=0.2,seed=113)
print('X Train Shape is : ' , x_train.shape)
print('X Train  is : ' , x_train[5])
print('---------------------------------------- ')
print('X Test Shape is : ' , x_test.shape)
print('X Test  is : ' , x_test[5])
print('---------------------------------------- ')
print('y Train Shape is : ' , y_train.shape)
print('y Train is : ' , y_train[5])
print('---------------------------------------- ')
print('y Test Shape is : ' , y_test.shape)
print('y Test  is : ' , y_test[5])
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data() # cifar10 is 60K random photos 50k of them for training and 10k for teating and label for 10 photo 

print('X Train Shape is : ' , x_train.shape)
print('X Train  is : ' , x_train[5])
print('---------------------------------------- ')
print('X Test Shape is : ' , x_test.shape)
print('X Test  is : ' , x_test[5])
print('---------------------------------------- ')
print('y Train Shape is : ' , y_train.shape)
print('y Train is : ' , y_train[5])
print('---------------------------------------- ')
print('y Test Shape is : ' , y_test.shape)
print('y Test  is : ' , y_test[5])
print('---------------------------------------- ')
print('All y is : ' , np.unique( y_train )) # get num of all unique photo because every photos reapated 10 times .

for i in range(10):
    plt.imshow(x_train[i])
    plt.show()
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data() # DataSet for 70K photos of numbers from 0 to 9 were writed by hands 60k training and 10K Teasting

print('X Train Shape is : ' , x_train.shape)
print('X Train  is : ' , x_train[5])
print('---------------------------------------- ')
print('X Test Shape is : ' , x_test.shape)
print('X Test  is : ' , x_test[5])
print('---------------------------------------- ')
print('y Train Shape is : ' , y_train.shape)
print('y Train is : ' , y_train[5])
print('---------------------------------------- ')
print('y Test Shape is : ' , y_test.shape)
print('y Test  is : ' , y_test[5])
print('---------------------------------------- ')
print('All y is : ' , np.unique( y_train ))

for i in range(10):
    #plt.imshow(x_train[i]) # photos will apear 3 color RGB ! 
    plt.imshow(x_train[i],cmap='Greys') # photo will apear balck and white only !
    plt.show()
    print("Number : ",y_train[i])
```
